```python
from PIL import Image # image open
import os # iterate on elements of a folder
import numpy as np
```

```python
# read the image dataset
dog_imgs = []
psize = 256
for filename in os.listdir(r'./Images/Yorkshire_terrier'): # iterate on filenames
    if filename.endswith(".jpg") or filename.endswith(".png"):
        img = Image.open('./Images/Yorkshire_terrier/' + filename)
        img = np.array(img.resize((psize,psize))).astype(np.float32).reshape(psize,psize,3) # w x h x rgb
        dog_imgs.append(img)
dog_imgs = np.asarray(dog_imgs)
# standard scale
mean, std = dog_imgs.mean(), dog_imgs.std()
dog_imgs = (dog_imgs - mean) / std
```

```python
# split to train-validation-test dataset
test_split = 0.1
valid_split = 0.1
v_index= int(dog_imgs.shape[0]*(1-valid_split-test_split))
t_index= int(dog_imgs.shape[0]*(1-test_split))

dog_imgs_test = dog_imgs[t_index:]
dog_imgs_valid = dog_imgs[v_index:t_index]
dog_imgs_train = dog_imgs[:v_index]
```

```python
from scipy.io import wavfile
# read the sound dataset
dog_barkings = []
for filename in os.listdir(r'./Sounds/dog_barking'): # iterate on filenames
    if filename.endswith(".wav"):
        samp_rate, data = wavfile.read('./Sounds/dog_barking/' + filename)
        dog_barkings.append(data)
dog_barkings = np.asarray(dog_barkings)
```

```python
# standard scale
mean, std = np.hstack(dog_barkings).mean(), np.hstack(dog_barkings).std()
dog_barkings = (dog_barkings - mean) / std
```

```python
# split to train-validation-test dataset
test_split = 0.1
valid_split = 0.1
v_index= int(dog_barkings.shape[0]*(1-valid_split-test_split))
t_index= int(dog_barkings.shape[0]*(1-test_split))

dog_barkings_test = dog_barkings[t_index:]
dog_barkings_valid = dog_barkings[v_index:t_index]
dog_barkings_train = dog_barkings[:v_index]
```
